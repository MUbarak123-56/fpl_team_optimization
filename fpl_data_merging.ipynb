```python
import pandas as pd
```

```python
df = pd.read_excel("data.xlsx")
```

```python
df2 = pd.read_csv("https://raw.githubusercontent.com/vaastav/Fantasy-Premier-League/master/data/2023-24/players_raw.csv")
```

```python
df2["name"] = df2["first_name"] + " " + df2["second_name"]
```

```python
df2.columns
```

```python
df2 = df2[["name", "news_added", "news", "points_per_game", "team", "form", "value_form", "value_season"]]
```

```python
df3 = pd.read_csv("https://raw.githubusercontent.com/vaastav/Fantasy-Premier-League/master/data/2023-24/teams.csv")
```

```python
df_new = df2.merge(df3, left_on = "team", right_on="id")
```

```python
df_new = df_new.rename(columns={"name_x":"name", "team":"team_code","name_y":"team", "form_x":"form"})
```

```python
df_new
```

```python
df_new=df_new[["name", "team", "news", "news_added", "points_per_game", "form", "value_form", "value_season"]]
```

```python
tot_df = df.merge(df_new, on=["name","team"])
```

```python
tot_df[tot_df["team"]=="Liverpool"]
```

```python
tot_df = tot_df[tot_df["news"].isna()].reset_index(drop=True)
```

```python
tot_df
```

```python
tot_df[tot_df["points_per_game_y"] > 0].reset_index(drop=True)
```

```python
latest_df = tot_df
```

```python
gk = latest_df[latest_df["position"]=="GK"].sort_values("points_per_game_y", ascending=False).reset_index(drop=True).head(5)
defenders = latest_df[latest_df["position"]=="DEF"].sort_values("points_per_game_y", ascending=False).reset_index(drop=True).head(15)
midfielders = latest_df[latest_df["position"]=="MID"].sort_values("points_per_game_y", ascending=False).reset_index(drop=True).head(15)
forwards = latest_df[latest_df["position"]=="FWD"].sort_values("points_per_game_y", ascending=False).reset_index(drop=True).head(15)

total_df = pd.concat([gk, defenders, midfielders, forwards], axis = 0).sort_values("position", ascending=False).reset_index(drop=True)
```
